```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import math
import scipy.stats as sps
from sympy import *
from matplotlib import cm
plt.style.use('ggplot')
%matplotlib inline
```

хотим посмотреть как меняется мощность критериев Пейджа и Фридмана при изменении параметров выборок и альтернатив

```python
N = np.arange(20,41,10)
K = np.arange(5,16,5)
B = np.arange(10,31,10)
Distrs = [sps.norm.rvs, sps.t.rvs, sps.uniform.rvs]
Params = [dict(), {'df':2}, {'loc':-1,'scale':2}] 
alpha = 0.05
```

```python
from statsmodels.sandbox.stats.multicomp import multipletests  # Holm method
from scipy.stats import friedmanchisquare  # Friedman test
```

```python
def get_stats(statf,n,k=100):
    stats = np.zeros(k)
    gens = sps.norm.rvs(size=(k,n))
    for i in range(k):
        stats[i] = statf(gens[i])[0]
    stats = np.sort(stats)
    return np.array(stats)

def get_mult_stats(statf, n, l, k=100):
    stats = np.zeros(k)
    gens = []
    for i in range(n):
        gens.append(sps.norm.rvs(size=(k,l)))
    for i in range(k):
        cur = []
        for j in range(n):
              cur.append(gens[j][i])
        stats[i] = statf(*cur)[0]
    stats = np.sort(stats)
    return np.array(stats)

def run_stats(x,statf,stats):
    st = statf(x)[0]
    p = np.sum((st > stats))/len(stats)
    return 2*min(p,1-p)  # two-tailed p-value

def run_stats_pointer(x,statf,stats):
    st = statf(*x)[0]
    p = np.sum((st > stats))/len(stats)
    return 2*min(p,1-p)  # two-tailed p-value
```

```python
def my_page_stat(*args):
    X = [x for x in args]
    Y = []
    for i in range(len(X[0])):
        Z = []
        for x in X:
            Z.append(x[i])
        Z = sorted(Z)
        Y.append(np.array(Z))
    R = np.zeros((len(X),len(X[0])))
    for i in range(len(X)):
        for j in range(len(X[0])):
            R[i][j] = np.sum(Z[j] < X[i][j])
    Rg = np.mean(R,axis=1)
    return np.sum(np.arange(len(X))*Rg),1  
# можно все считать в средних, так как при моделировании это не будет важно
```

```python
def calc_power(statf, n, k, b, distr, params, alpha, grid_model=10000, n_tests=1000):
    stats = get_mult_stats(statf, k, n, grid_model)
    crit = (lambda s: run_stats_pointer(s, statf, stats))
    res = np.zeros((2,2))
    params['size'] = (n_tests, k, n)
    data = distr(**params)
    for t in range(n_tests):
        cur = data[t]
        c = crit(cur)
        if c > alpha:
            res[1][1] += 1
        else:
            res[0][0] += 1
        for i in range(k):
            data[t][i] += b*(i+1)
        cur = data[t]
        c = crit(cur)
        if c > alpha:
            res[0][1] += 1
        else:
            res[1][0] += 1
    return  res[0][0]/(res[1][1] + res[0][0]), res[1][0]/(res[1][0] + res[0][1])
```

```python
%%time
e1 = dict()
e2 = dict()
for statf in [friedmanchisquare,my_page_stat]:
    for n in N:
        for k in K:
            for b in B:
                for i in range(len(Distrs)):
                    print('crit=' + statf.__name__
                        , 'n='+str(n)
                        ,'k='+str(k)
                        , 'b='+str(b)
                        , 'alpha=0.05'
                        , 'distr='+str(i)
                        )
                    cur_e1, cur_e2 = calc_power(statf
                                                , n
                                                , k
                                                , b
                                                , Distrs[i]
                                                , Params[i]
                                                , alpha
                                                , 1000
                                                , 1000)
                    e1[(n,k,b,statf.__name__,i)] = cur_e1
                    e2[(n,k,b,statf.__name__,i)] = cur_e2
                    print('error1:' + str(cur_e1) + ', power:' + str(cur_e2))
```

Уже заметно, что на мощность критерия Фридмана не влияет изменение параметров и его мощность близка к единице против данной альтернативы. Посмотим на результаты критерия Пэйджа:

```python
data_x = N
plt.figure(figsize=(15,10))
plt.title('Norm')
for k in K:
    for b in B:
        data_y = [e2[(n,k,b,'my_page_stat',0)] for n in N]
        plt.plot(data_x, data_y,label='k='+str(k)+',b='+str(b))
plt.legend(loc='best')
plt.show()
data_x = N
plt.figure(figsize=(15,10))
plt.title('Student_2')
for k in K:
    for b in B:
        data_y = [e2[(n,k,b,'my_page_stat',1)] for n in N]
        plt.plot(data_x, data_y,label='k='+str(k)+',b='+str(b))
plt.legend(loc='best')
plt.show()
data_x = N
plt.figure(figsize=(15,10))
plt.title('Uniform_[-1,1]')
for k in K:
    for b in B:
        data_y = [e2[(n,k,b,'my_page_stat',2)] for n in N]
        plt.plot(data_x, data_y,label='k='+str(k)+',b='+str(b))
plt.legend(loc='best')
plt.show()
```

Видно, что мощность падает с увеличением числа выборок на всех рассматриваемых распределениях. При этом параметр $\beta$ влияет незначительно.
