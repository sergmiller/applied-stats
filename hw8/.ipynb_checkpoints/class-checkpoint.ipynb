```r
data = read.csv('good_data.csv',sep = ',',header = TRUE)
```

```r
head(data)
```

```r
## 75% of the sample size
smp_size <- floor(0.75 * nrow(data))

## set the seed to make your partition reproductible
set.seed(123)
train_ind <- sample(seq_len(nrow(data)), size = smp_size)

train <- data[train_ind, ]
test <- data[-train_ind, ]
```

```r
fit<-lm(MPG ~ VOL + HP + SP, data=train)
pred<-as.numeric(predict(fit, test))
```

```r
cor(method = 'pearson', pred, test$MPG)  # R^2
```

```r
fit$coefficients
```

```r
ost = pred - test$MPG
```

```r
std = sd(ost)
```

```r
ALPHA = 0.05
```

```r
inter = matrix(rep(0, 2*length(pred)),nrow=2)
for(i in 1:length(pred)) {
    inter[1,i] = pred[i] + qnorm(1- ALPHA/2,mean = 0, sd=std)
    inter[2,i] = pred[i] - qnorm(1- ALPHA/2,mean = 0, sd=std)
}
inter
```

```r
plot(pred)
lines( 
      pred, 
      col = "red", lwd = 2)

lines(inter[1,],col = "green", lwd = 2)

lines( inter[2,],col = "blue", lwd = 2)
```
