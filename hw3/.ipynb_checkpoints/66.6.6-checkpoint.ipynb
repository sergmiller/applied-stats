Нужно надо найти такой n, что распределение статистики Смирнова не отличается от распределения Колмогорова на уровне $\alpha = 0.05$. 

Так как статистика не зависит от распределения при верности гипотезы, то для моделирования будем использовать выборки равномерного распределения (100 шт для каждого n).

Как понять, что минимальный нужный n найден? Просто смоделируем выбороки, посчитаем статистику и методом множественной проверки проверим что статистика из распределения Колмогорова на уровне $FWER \leq \alpha = 0.05$ (например критерием Андерсона-Дарлинга).

Так как статистики будут независимыми то оптимальее всего для множественной проверки использовать нисходящий метод Шидака-Холма.

```python
from statsmodels.sandbox.stats.multicomp import multipletests  # Holm method
from skgof import ad_test  # Anderson-Darling test
```

```python
def EMP(X):
    @np.vectorize
    def f(y):
        return np.sum(X < y)/len(X)
    return f

def get_smirn(X,Y):
    X = np.array(X)
    Y = np.array(Y)
    X = np.sort(X)
    Y = np.sort(Y)
    d = np.arange(len(X)) + 1
    maxx = max(np.max(d/len(X) - EMP(X)(Y)),np.max(d/len(X) - EMP(Y)(X)))     
    return (len(X)/2)**0.5 * maxx
```

При n >= 25 можно использовать для проверки критерий Андерсона-Дарлинга, при n < 25 не будем учитывать результаты

```python
%%time

K = 100
p_values = []

for N in range(1,500):
    data = sps.uniform.rvs(size=(K,2,N))
    stats = [get_smirn(x[0],x[1]) for x in data]
    p_values.append(ad_test(stats,sps.kstwobign)[1])
    
a = 0.05
reject = multipletests(pvals=p_values, method='holm-sidak',alpha=a)[0]
print('for alpha = ' + str(a) + ' : ' + str(np.sum(reject)) + ' simple hypotheses are rejected:')
print('p_values: ' + str(p_values))
print('reject_bits: ' + str(reject))
print()
```

```python
plt.figure(figsize=(15,5))
plt.plot(np.arange(1,500),reject,'*',alpha=0.5)
```

Видно что при примерно n > 30-40 остаются только единичные случаи несоответствия гипотезе. Но так как проверка множетсвенная, то эти редкие несоответствия гипотезе лучше учесть и взять последнее n где гипотеза отвергается:

```python
argm = 0
for i in range(499):
    if reject[i] == 1:
        argm = i+1
print(argm)
```
