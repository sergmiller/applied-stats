```r
N <- 400
q <- 100
```

```r
x_train <- data[1:N, 1:2]
x_test <- data[N+1:500, 1:2]
y_train <- data$V3[1:N]
y_test <- data$V3[N+1:500]

train <- data[1:N,]
test <- data[N+1:500,]
```

```r
# install.packages('bbemkr')
library('bbemkr')
library('np')
```

```r
bw <- npregbw(V3 ~ V1 + V2, data = train)
np_model <- npreg(bws = bw, gradients = TRUE)
```

```r
p <- lm(V3 ~ V1 + V2, data = train)
p_pred = c(predict(p, x_test))[1:100]

np_pred = predict(np_model, new_data=x_test)
```

```r
p_pred
```

```r
mean(abs(p_pred - y_test)/((abs(p_pred) + abs(y_test))/2))
```
